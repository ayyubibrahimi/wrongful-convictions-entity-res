```python
import pandas as pd
```

```python
dfa = pd.read_csv("../data/contacts.csv")

dfa = dfa[["RelationshipDescription", "Contact"]]

dfa = dfa.rename(columns={"RelationshipDescription": "title", "Contact": "officer_name"})

dfa = dfa[dfa.title.str.lower().str.strip().str.contains("police", "detective") ]

dfa = dfa[["officer_name"]]

dfa
```

```python
dfb = pd.read_csv("../../preprocessing/data/output/clean.csv")

dfz = dfb.copy()

dfb = dfb[["officer_name"]]
```

```python
dfc = pd.read_csv("../data/personnel.csv")

dfc.loc[:, "officer_name"] = dfc.first_name.str.cat(dfc.last_name, sep=" ").str.title() 

dfc = dfc[["officer_name", "agency"]] 

dfc
```

```python
def find_perfect_matches(dfa, dfb, dfc):
    merged_df = pd.merge(dfa, dfb, on='officer_name', how='inner')
    merged_df = pd.merge(merged_df, dfc, on='officer_name', how='inner')

    return merged_df

perf_matches = find_perfect_matches(dfa, dfb, dfc)  

perf_matches = perf_matches.drop_duplicates(subset=["officer_name"])

perf_matches
```
